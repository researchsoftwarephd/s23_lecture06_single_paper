```python
from pyclugen import clugen                 # Function to create data
from sklearn.cluster import KMeans          # Class to perform k-Means clustering
from sklearn.metrics import v_measure_score # Function to assess clustering quality
import matplotlib.pyplot as plt             # Module to plot stuff
from numpy.random import default_rng as rng # Get NumPy's default random number generator
```

```python
xy = clugen(2, 4, 1000, [1, 1], 0.5, [10, 10], 20, 2.5, 1.5, rng=rng(1234)) # Generate some data...
plt.scatter(xy.points[:, 0], xy.points[:, 1], c=xy.clusters)                # ...and show it
```

```python
kmf = KMeans(n_clusters=4, random_state=345).fit(xy.points)  # Automatically group data with k-means
plt.scatter(xy.points[:, 0], xy.points[:, 1], c=kmf.labels_) # Show how k-means grouped our data
```

```python
v_measure_score(xy.clusters, kmf.labels_) # Get clustering quality, between 0 and 1
```
